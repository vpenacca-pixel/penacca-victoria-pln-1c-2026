## Introducción

Esta guía te acompaña desde la configuración inicial de Git y GitHub hasta la publicación de tu primer repositorio de cursada.

Está pensada para estudiantes de **Procesamiento Digital de Imágenes (PDI)** y **Procesamiento del Lenguaje Natural (PLN)** que todavía no trabajaron con control de versiones o que necesitan ordenar su flujo de trabajo.

No necesitás experiencia previa. El objetivo no es solo subir archivos: vas a construir un repositorio claro, consistente y útil para seguir la materia, entregar actividades y conservar un registro de tu producción a lo largo del cuatrimestre.


## Objetivos de aprendizaje

Al terminar esta guía vas a poder:

1. Diferenciar qué hace **Git** y qué hace **GitHub**.
2. Configurar tu identidad en Git para empezar a versionar archivos.
3. Crear un repositorio en GitHub con la nomenclatura correcta de la materia.
4. Organizar la estructura semanal del repositorio con el formato institucional.
5. Subir notebooks, imágenes, PDF y diapositivas sin desordenar la cursada.
6. Compartir enlaces de entrega desde GitHub sin comprimir archivos ni habilitar permisos extra en Google Colab.
7. Resolver los problemas más comunes del flujo básico de trabajo.


## Parte 1: Qué son Git y GitHub

**Git** es un sistema de control de versiones. Te permite registrar cambios en tus archivos, volver a estados anteriores y trabajar de manera ordenada.

**GitHub** es una plataforma que aloja repositorios Git en la web. Funciona como espacio de publicación, respaldo y entrega.

En estas materias, tu repositorio personal cumple varias funciones al mismo tiempo:

- organiza tus materiales semana a semana;
- conserva notebooks y otros archivos importantes;
- permite al docente revisar tu trabajo;
- simplifica las entregas, porque compartís enlaces ya publicados en GitHub.

En otras palabras: tu repositorio no es un depósito improvisado de archivos. Es tu carpeta de cursada versionada.


## Parte 2: Convención institucional para nombrar el repositorio

El nombre del repositorio debe respetar esta estructura:

```text
apellido-nombre-materia-cuatrimestre-anio
```

Ejemplos correctos:

- `perez-ana-pdi-1c-2026`
- `gomez-juan-pln-2c-2027`
- `lopez-maria-pdi-1c-2026`

Componentes del nombre:

| Parte | Significado | Ejemplo |
|---|---|---|
| `apellido` | tu apellido | `perez` |
| `nombre` | tu nombre | `ana` |
| `PLN` o `PDI` | materia | `pdi` |
| `1c` o `2c` | cuatrimestre | `1c` |
| `año` | año lectivo | `2026` |

Reglas prácticas:

- usá **minúsculas**;
- usá **guiones medios** (`-`) en lugar de espacios;
- no uses tildes;
- si tu nombre o apellido tiene `ñ`, reemplazala por `n` simple;
- mantené siempre el mismo criterio de escritura.


## Parte 3: Estructura interna obligatoria del repositorio

Dentro del repositorio, la organización semanal debe seguir un orden estricto.

Cada semana se nombra con tres dígitos:

```text
001, 002, 003, 004, ...
```

Dentro de cada semana van siempre estas tres carpetas:

```text
001 - TEO
002 - PRA
003 - LAB
```

Significado de cada carpeta:

| Carpeta | Uso |
|---|---|
| `001 - TEO` | cuadernos y materiales usados en teóricos |
| `002 - PRA` | cuadernos y materiales usados en prácticos |
| `003 - LAB` | ejercicios de laboratorio que el docente solicita para seguimiento |

Ejemplo de estructura:

```text
perez-ana-pdi-1c-2026/
├── 001/
│   ├── 001 - TEO/
│   ├── 002 - PRA/
│   └── 003 - LAB/
├── 002/
│   ├── 001 - TEO/
│   ├── 002 - PRA/
│   └── 003 - LAB/
└── 003/
    ├── 001 - TEO/
    ├── 002 - PRA/
    └── 003 - LAB/
```

No hace falta definir desde ahora cuántas semanas tendrá el cuatrimestre. Lo importante es que mantengas siempre el mismo patrón y lo repliques a medida que avance la cursada.


## Parte 3.1: Regla clave sobre carpetas vacías

Git **no versiona carpetas vacías**. Esto significa que, si creás una carpeta pero no guardás ningún archivo adentro, esa carpeta no aparecerá en GitHub.

Tenés dos caminos válidos:

1. Crear las carpetas cuando ya tengas material para guardar.
2. Dejar la estructura preparada y agregar un archivo de apoyo, por ejemplo `README.md` o `.gitkeep`.

Si querés dejar la semana `001` armada desde el inicio, podés usar este esquema:

```bash
mkdir -p "001/001 - TEO" "001/002 - PRA" "001/003 - LAB"
touch "001/001 - TEO/.gitkeep" "001/002 - PRA/.gitkeep" "001/003 - LAB/.gitkeep"
```

Si trabajás desde el explorador de archivos y no desde la terminal, el criterio es el mismo: cada carpeta vacía necesita al menos un archivo para quedar registrada en Git.


## Parte 4: Preparación del entorno

Antes de crear tu repositorio, prepará estas tres cosas:

1. **Instalar Git** desde [git-scm.com](https://git-scm.com/downloads).
2. **Crear tu cuenta en GitHub** desde [github.com/signup](https://github.com/signup).
3. **Elegir una ubicación local de trabajo** en tu computadora, por ejemplo una carpeta llamada `Repositorios` o `Cursada`.

Recomendación de trabajo:

- si usás Windows, instalá Git con **Git Bash** y **Git Credential Manager**;
- usá el mismo correo en Git y en GitHub siempre que sea posible;
- para esta cursada, conviene que tu repositorio sea **público**, salvo indicación docente en contrario. De ese modo, el enlace de entrega funciona sin pedir permisos adicionales.

Para verificar que Git quedó instalado correctamente, abrí la terminal y ejecutá:

```bash
git --version
```


## Parte 4.1: Configuración de identidad en Git

Git necesita saber quién registra los cambios. Configuralo una sola vez en tu computadora.

Reemplazá los datos de ejemplo por los tuyos:

```bash
git config --global user.name "Tu Nombre y Apellido"
git config --global user.email "tu-correo@ejemplo.com"
git config --global init.defaultBranch main
```

Para revisar la configuración actual:

```bash
git config --global --list
```

Si en algún momento Git te muestra un error parecido a `Author identity unknown`, casi siempre se resuelve con esta configuración.


## Parte 5: Crear el repositorio en GitHub

La ruta recomendada para esta guía es crear primero el repositorio en GitHub y después clonarlo a tu computadora.

Pasos:

1. Entrá a GitHub.
2. Hacé clic en **New repository**.
3. Escribí el nombre exacto con la convención institucional.
4. Elegí visibilidad **Public** si no te indicaron otra cosa.
5. Marcá la opción **Add a README file** para arrancar con una rama `main` ya creada.
6. Creá el repositorio.

Ejemplo:

```text
perez-ana-pln-1c-2026
```

Antes de continuar, revisá dos veces el nombre. Corregirlo al principio es muy fácil; corregirlo tarde genera trabajo innecesario.


## Parte 6: Clonar el repositorio en tu computadora

Una vez creado el repositorio remoto, copiás su contenido a tu máquina local con `git clone`.

En GitHub:

1. Abrí tu repositorio.
2. Hacé clic en **Code**.
3. Copiá la URL HTTPS.

En la terminal, ubicate en la carpeta donde querés trabajar y ejecutá:

```bash
git clone https://github.com/tu-usuario/perez-ana-pln-1c-2026.git
cd perez-ana-pln-1c-2026
```

A partir de ese momento, todo lo que guardes dentro de esa carpeta puede versionarse con Git.

Si ya habías empezado a trabajar en una carpeta local sin clonar el repositorio, también es posible vincularla después. Sin embargo, para comenzar, este flujo suele ser el más claro y el menos propenso a errores.


## Parte 7: Crear la estructura inicial de semanas y subcarpetas

Ya dentro del repositorio clonado, creá la estructura de trabajo. Podés empezar por la semana `001` y después replicar el patrón.

Ejemplo en Git Bash:

```bash
mkdir -p "001/001 - TEO" "001/002 - PRA" "001/003 - LAB"
touch "001/001 - TEO/.gitkeep" "001/002 - PRA/.gitkeep" "001/003 - LAB/.gitkeep"
```

Si querés dejar también la semana `002` preparada:

```bash
mkdir -p "002/001 - TEO" "002/002 - PRA" "002/003 - LAB"
touch "002/001 - TEO/.gitkeep" "002/002 - PRA/.gitkeep" "002/003 - LAB/.gitkeep"
```

Alternativa manual:

- creá las carpetas desde el explorador de archivos;
- agregá después al menos un archivo dentro de cada carpeta que quieras publicar;
- evitá nombres alternativos como `Teorico`, `Practica`, `Lab 1` o `Semana 1`. En esta cursada la estructura debe ser consistente.


## Parte 8: Primer flujo completo de Git

Una vez creada la estructura inicial, llega el momento de registrar y publicar los cambios.

Ejecutá estos comandos dentro del repositorio:

```bash
git status
git add .
git commit -m "Crear estructura inicial del repositorio"
git push origin main
```

Qué hace cada paso:

- `git status` muestra qué cambió;
- `git add .` prepara los archivos para el commit;
- `git commit -m "..."` guarda un punto de control local con mensaje;
- `git push origin main` publica esos cambios en GitHub.

Cuando termines, actualizá la página del repositorio en GitHub y verificá que la estructura ya aparezca publicada.


## Parte 9: Flujo cotidiano de trabajo durante la cursada

A partir de la primera publicación, el flujo habitual será este:

```bash
git pull origin main
# editar o agregar archivos
git status
git add .
git commit -m "Mensaje descriptivo"
git push origin main
```

Regla de trabajo recomendada:

- antes de empezar una sesión, ejecutá `git pull origin main`;
- trabajá dentro de la carpeta correcta de la semana;
- usá mensajes de commit claros y concretos.

Ejemplos de mensajes útiles:

- `Agregar notebook de teórico de la semana 001`
- `Subir PDF del práctico 002`
- `Actualizar ejercicios de laboratorio de la semana 003`
- `Corregir imágenes de apoyo del teórico 004`


## Parte 10: Cómo subir notebooks, imágenes, PDF y diapositivas

Este repositorio no guarda solo código. También funciona como carpeta ordenada de materiales de cursada.

Archivos que vas a usar con frecuencia:

- notebooks de Jupyter o Colab: `.ipynb`
- imágenes: `.png`, `.jpg`, `.jpeg`
- documentos: `.pdf`
- diapositivas: `.pdf`, `.pptx`

Recomendaciones por tipo de archivo:

### Notebooks

- si trabajaste en Google Colab, descargá el archivo `.ipynb` y guardalo en la carpeta correcta del repositorio;
- revisá que el nombre del archivo sea claro y consistente;
- hacé `add`, `commit` y `push` igual que con cualquier otro material.

### Imágenes

- guardá capturas, esquemas o figuras dentro de la carpeta de la semana correspondiente;
- evitá dejar imágenes sueltas fuera de contexto.

### PDF

- subí guías, apuntes exportados, informes o consignas cuando corresponda;
- el formato PDF suele verse muy bien en GitHub.

### Diapositivas

- podés subir archivos `.pptx`, pero la vista previa puede variar según el navegador;
- si querés asegurar una lectura rápida, conviene subir también una versión en PDF.

Si preferís, algunos archivos también pueden subirse desde la web de GitHub con **Add file > Upload files**. Sin embargo, para mantener un flujo consistente, en esta guía priorizamos el trabajo con Git.


## Parte 11: Entregas por formulario y uso de enlaces

Cuando el docente te pida enviar un cuaderno o material por formulario, la entrega recomendada es el **enlace del archivo ya publicado en tu repositorio**.

Esto evita varios problemas frecuentes:

- no tenés que comprimir archivos;
- no dependés de permisos manuales de Google Colab;
- el material queda respaldado en un entorno ordenado;
- el docente puede revisar exactamente la versión que publicaste.

Procedimiento sugerido:

1. Abrí el archivo en GitHub una vez hecho el `push`.
2. Verificá que cargue correctamente.
3. Copiá la URL del navegador.
4. Pegá ese enlace en el formulario.

Si el formulario pide una entrega puntual, enviá el enlace al archivo. Si pide una revisión más amplia, podés enviar el enlace a la carpeta semanal o al repositorio completo, según corresponda.


## Parte 12: Ramas y pull requests en versión simplificada

En un repositorio personal de cursada, la mayor parte del tiempo vas a trabajar directamente sobre `main`. Eso alcanza para el flujo básico de la materia.

Si en algún momento el docente te pide trabajar con una rama, el esquema básico es este:

```bash
git switch -c correccion-lab-001
# realizar cambios
git add .
git commit -m "Corregir ejercicio de laboratorio 001"
git push -u origin correccion-lab-001
```

Después podés abrir un **pull request (PR)** en GitHub. Un PR es una solicitud para incorporar cambios de una rama a otra, generalmente hacia `main`.

Para esta guía, alcanza con que recuerdes dos ideas:

- `main` es tu rama principal;
- una rama alternativa sirve para aislar cambios cuando te lo pidan de manera explícita.


## Parte 13: Troubleshooting básico

A continuación tenés algunos problemas frecuentes y su resolución.

### 1. `git` no se reconoce en la terminal

**Causa probable:** Git no se instaló correctamente o la terminal no se reinició.

**Solución:** reinstalá Git si hace falta y volvé a abrir la terminal. Luego ejecutá:

```bash
git --version
```

### 2. `Author identity unknown`

**Causa probable:** Git no tiene configurado tu nombre y tu correo.

**Solución:**

```bash
git config --global user.name "Tu Nombre y Apellido"
git config --global user.email "tu-correo@ejemplo.com"
```

### 3. `fatal: Authentication failed`

**Causa probable:** la autenticación con GitHub falló.

**Solución:**

- intentá iniciar sesión otra vez desde el flujo de autenticación del sistema;
- si GitHub te pide contraseña para `push`, no uses la contraseña común de tu cuenta: usá un **Personal Access Token** o el método de autenticación que GitHub te indique.

### 4. `! [rejected] main -> main (non-fast-forward)`

**Causa probable:** el repositorio remoto tiene cambios que tu copia local todavía no incorporó.

**Solución:**

```bash
git pull origin main
git push origin main
```

### 5. La carpeta aparece en tu computadora pero no en GitHub

**Causa probable:** la carpeta estaba vacía o faltaron pasos del flujo Git.

**Solución:**

- agregá al menos un archivo dentro de la carpeta;
- ejecutá `git status` para revisar que Git detecte el cambio;
- después hacé `git add .`, `git commit` y `git push`.

### 6. El archivo existe localmente pero no se ve publicado

**Causa probable:** faltaron uno o más pasos entre `add`, `commit` y `push`.

**Solución:**

```bash
git status
git add .
git commit -m "Subir archivo faltante"
git push origin main
```

### 7. El enlace de Colab pide permisos

**Causa probable:** estás compartiendo el enlace de Colab y no el de GitHub, o el archivo no quedó publicado en un repositorio accesible.

**Solución:**

- priorizá el enlace del archivo publicado en GitHub;
- usá un repositorio público si no te indicaron otra cosa.

### 8. El repositorio tiene un nombre incorrecto

**Causa probable:** al crearlo no seguiste la convención institucional.

**Solución recomendada:** corregilo cuanto antes. Si recién empezaste, conviene crear el repositorio correctamente y mover el contenido. Si ya avanzaste bastante, también podés renombrarlo desde GitHub y actualizar la URL remota local.


## Parte 14: Ejercicios prácticos

### Ejercicio 1: Crear el repositorio correcto

1. Creá un repositorio en GitHub con el formato institucional.
2. Verificá que tenga nombre correcto, visibilidad pública y `README.md` inicial.

**Resultado esperado:** el repositorio existe en GitHub y respeta la nomenclatura de la cursada.

---

### Ejercicio 2: Clonar y preparar la semana 001

1. Cloná el repositorio en tu computadora.
2. Creá la carpeta `001`.
3. Dentro de `001`, creá `001 - TEO`, `002 - PRA` y `003 - LAB`.
4. Agregá archivos de apoyo si las carpetas quedan vacías.
5. Publicá los cambios con `add`, `commit` y `push`.

**Resultado esperado:** la estructura semanal aparece en GitHub.

---

### Ejercicio 3: Subir un notebook

1. Guardá un notebook en `001/001 - TEO/` o `001/002 - PRA/`.
2. Publicalo con Git.
3. Verificá que puedas abrirlo desde GitHub.

**Resultado esperado:** el notebook se ve en el repositorio y tiene una URL propia.

---

### Ejercicio 4: Subir un material complementario

1. Agregá una imagen o un PDF en la carpeta correspondiente.
2. Publicá los cambios.
3. Comprobá que el archivo se abra correctamente desde GitHub.

**Resultado esperado:** el repositorio contiene más de un tipo de material, no solo notebooks.

---

### Ejercicio 5: Preparar una entrega por enlace

1. Abrí en GitHub uno de tus archivos publicados.
2. Copiá la URL del navegador.
3. Pegala en un documento o formulario de prueba.

**Resultado esperado:** podés entregar un enlace funcional sin comprimir archivos ni gestionar permisos extra.


## Cierre y lista de verificación final

Antes de dar por terminado tu repositorio inicial, revisá esta lista:

- el nombre del repositorio cumple la convención institucional;
- la materia (`pdi` o `pln`), el cuatrimestre y el año son correctos;
- la estructura semanal usa `001`, `002`, `003`, ...;
- dentro de cada semana aparecen `001 - TEO`, `002 - PRA` y `003 - LAB`;
- los archivos están ubicados en la carpeta correcta;
- hiciste `git add`, `git commit` y `git push`;
- verificaste en GitHub que los materiales realmente se publicaron;
- sabés copiar el enlace de un archivo para una entrega.

Recursos útiles para profundizar:

- [Documentación oficial de Git](https://git-scm.com/doc)
- [Documentación oficial de GitHub](https://docs.github.com/)
- [GitHub Skills](https://skills.github.com/)

A partir de este punto, tu repositorio personal deja de ser una carpeta improvisada y pasa a ser una base ordenada para toda la cursada.
