## 2. Tipos de Datos: ¿Qué hay dentro de las cajas?

No todos los datos son iguales. Python distingue principalmente entre:

* **String (`str`):** Texto, siempre entre comillas (`" "`). ¡El tipo de dato más importante para nosotros en PLN!
* **Integer (`int`):** Números enteros (ej: `10`, `-5`).
* **Float (`float`):** Números con decimales (ej: `3.14`).
* **Boolean (`bool`):** Representa la verdad. Solo puede ser `True` o `False`.

Para saber qué tipo de dato tiene una variable, usamos la función `type()`.

````python
# Variable de tipo String
oracion = "El análisis de texto es fascinante."

# Variable de tipo Integer
cantidad_de_palabras = 6

# Variable de tipo Boolean
es_interesante = False

variable_float = 3.14

# Usemos type() para verificar
print(type(oracion))
print(type(cantidad_de_palabras))
print(type(es_interesante))
print(type(variable_float))
````

> **Nota importante:** ¿Por qué nos importa el tipo de dato? Porque define qué operaciones podemos hacer. Por ejemplo, podemos `sumar` dos números, pero no podemos `dividir` un texto. `type()` es nuestro mejor amigo para no cometer errores.

## 3. Operadores: Haciendo Cosas con los Datos

Podemos combinar y comparar nuestras variables usando operadores.

* **Operador `+` con texto:** Une (concatena) dos strings.
* **Operadores de comparación (`==`, `!=`):** Preguntan si dos valores son iguales o diferentes. ¡Siempre devuelven un Booleano (`True` o `False`)!

### El operador `in`: pertenencia

El operador `in` verifica si un elemento está dentro de una colección o de un string. Devuelve siempre un booleano (`True` o `False`).

Es uno de los operadores más útiles en PLN: permite verificar si una palabra aparece en un texto, si un término pertenece a un vocabulario o si un patrón está presente en una oración.

````python
# Con listas: ¿está el elemento en la colección?
vocabulario = ["python", "nlp", "texto", "corpus"]
print("¿'nlp' está en el vocabulario?", "nlp" in vocabulario)
print("¿'gramática' está en el vocabulario?", "gramática" in vocabulario)

# Con strings: ¿está la subcadena dentro del texto?
oracion = "El procesamiento del lenguaje natural es fascinante"
print("¿'lenguaje' aparece en la oración?", "lenguaje" in oracion)
print("¿'Python' aparece en la oración?", "Python" in oracion)
````

## 5. Estructuras de Datos: Organizando nuestra Información

En PLN, necesitamos manejar colecciones de datos. Python nos ofrece varias formas de hacerlo:

* **Listas (`list`):** Colecciones ordenadas y modificables de elementos.
* **Tuplas (`tuple`):** Colecciones ordenadas pero INMODIFICABLES de elementos.
* **Diccionarios (`dict`):** Colecciones de pares clave-valor (como un diccionario real).

````python
# Listas (pueden modificarse)
palabras = ['hola', 'mundo', 'nlp', 'python']
print("Lista original:", palabras)
````

````python
# Agregar elemento
palabras.append('texto')
print("Después de append:", palabras)
````

````python
# Remover elemento
palabras.remove('mundo')
print("Después de remove:", palabras)
````

````python
# Acceder por índice
print("Primera palabra:", palabras[0])
print("Última palabra:", palabras[-1])
````

````python
# Tuplas (NO pueden modificarse)
coordenadas = (4, 5)
colores = ('rojo', 'verde', 'azul')

print("Tupla de coordenadas:", coordenadas)
print("Primer color:", colores[0])
print("Segundo color:", colores[1])
````

## 6. Bucles: Recorriendo Colecciones

Un **bucle** es una instrucción que repite un bloque de código para cada elemento de una colección. En PLN los usamos constantemente: para procesar listas de palabras, recorrer documentos de un corpus o aplicar operaciones a cada oración de un texto.

El bucle `for` itera sobre una colección, tomando de a un elemento por vez:

```python
for elemento in coleccion:
    # Este bloque se repite una vez por cada elemento
    hacer_algo(elemento)
```

La **indentación** (el sangrado con espacios) es fundamental: define qué líneas pertenecen al cuerpo del bucle.

````python
# Bucles para procesar listas
texto = "Python es un lenguaje de programación ampliamente utilizado en PLN"

print("Palabras largas en el texto:")
for palabra in texto.split():
    if len(palabra) > 5:  # Palabras de más de 5 letras
        print(f"- {palabra} ({len(palabra)} letras)")
````

- *"¿Cómo ordenar un diccionario por sus valores de mayor a menor?"*

````python
# Ejercicio: Contador de palabras
texto = "Python es un lenguaje de programación ampliamente utilizado en el procesamiento del lenguaje natural"

# Tu código acá:
# 1. Convertir texto a minúsculas
# 2. Dividir en palabras
# 3. Contar frecuencia de cada palabra

print("¡Intenta resolver este ejercicio!")
````

````python
# Ejercicio: Contador de palabras. Mi respuesta a la pregunta del ejercicio es la siguiente:
texto = "Python es un lenguaje de programación ampliamente utilizado en el procesamiento del lenguaje natural"

# 1. Convertir texto a minúsculas
texto_min = texto.lower()

# 2. Dividir en palabras
palabras = texto_min.split()

# 3. Contar frecuencia de cada palabra
frecuencias = {}

for p in palabras:
    if p in frecuencias:
        frecuencias[p] += 1  # Si ya existe, le sumo 1
    else:
        frecuencias[p] = 1   # Si es nueva, empiezo en 1


# Mostrar el resultado encolumnado para mejor visualización

print(f"{'Palabra':<15} | {'Frecuencia'}")
print("-" * 30)

for palabra, conteo in frecuencias.items():
    print(f"{palabra:<15} | {conteo}")
````

## Glosario Rápido — Clase 1

* **Variable**: Un nombre que usamos para guardar un valor en la memoria del programa.
* **String (`str`)**: Tipo de dato que representa texto. Se define entre comillas (`"hola"`).
* **Integer (`int`)**: Tipo de dato para números enteros, sin decimales (ej: `10`, `-3`).
* **Float (`float`)**: Tipo de dato para números con decimales (ej: `9.5`, `3.14`). Muy usado en PLN para probabilidades y pesos de modelos.
* **Boolean (`bool`)**: Tipo de dato con solo dos valores posibles: `True` o `False`.
* **Operador**: Símbolo que realiza una acción sobre datos. Ejemplos: `+` para unir texto, `==` para comparar igualdad, `in` para verificar pertenencia.
* **Operador `in`**: Verifica si un elemento está dentro de una colección o cadena. Devuelve `True` o `False`.
* **Condicional (`if`, `elif`, `else`)**: Estructura que permite ejecutar distintos bloques de código según se cumplan o no ciertas condiciones.
* **Bucle (`for`)**: Estructura que repite un bloque de código para cada elemento de una colección.
* **Lista (`list`)**: Colección ordenada y modificable de elementos.
* **Tupla (`tuple`)**: Colección ordenada e inmutable de elementos.
* **Diccionario (`dict`)**: Colección de pares clave-valor. Permite acceso eficiente a valores por nombre.
* **Indentación**: Los espacios al inicio de una línea. En Python, define qué código pertenece a un bloque (`if`, `for`, etc.).
* **Función**: Un bloque de código con nombre que realiza una tarea específica (ej: `print()`, `len()`, `type()`).
