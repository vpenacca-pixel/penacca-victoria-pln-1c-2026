## 2. Funciones: Creacion de Herramientas a Medida

Ya sabemos crear funciones, pero ahora vamos a usarlas como lo hacen los profesionales: para encapsular una lógica específica y hacer nuestro código más limpio y reutilizable. Una buena función hace **una sola cosa y la hace bien**.

Vamos a crear una función que limpie un texto y otra que, usando la primera, cuente las palabras.

```python
# Función 1: Especializada en limpiar texto
def limpiar_texto(texto):
    """Limpia un texto removiendo puntuación común, incluidas marcas del español."""
    # Nota: en español necesitamos incluir ¿ y ¡ además de los signos estándar.
    # También eliminamos comillas y paréntesis para obtener palabras limpias.
    caracteres_a_eliminar = ['.', ',', '!', '?', ';', ':', '¿', '¡', '"', "'", '(', ')']
    texto_limpio = texto
    for char in caracteres_a_eliminar:
        texto_limpio = texto_limpio.replace(char, '')
    return texto_limpio.lower()  # Devuelve el texto limpio y en minúsculas
```

```python
#Agregué un print para ver lo que ejecutaba
print(f"Texto limpio: '{limpiar_texto(texto)}'")
```

```python
# Función 2: Especializada en contar palabras
def contar_palabras(texto):
    """Usa limpiar_texto() para procesar el texto y luego cuenta las palabras."""
    texto_procesado = limpiar_texto(texto) # ¡Reutilizamos nuestra primera función!
    palabras = texto_procesado.split()
    return len(palabras)
```

```python
#Agregué un print para ver lo que ejecutaba
cantidad = contar_palabras(texto)
print(f"Cantidad de palabras: {cantidad}")
```

```python
# --- Uso de las funciones ---
mi_texto = "¡Hola, mundo! ¿Listo/a para aprender sobre Procesamiento del Lenguaje Natural?"
cantidad = contar_palabras(mi_texto)

print(f"El texto original era: '{mi_texto}'")
print(f"Después de limpiarlo y contarlo, tiene {cantidad} palabras.")
```

## 3. Programación Orientada a Objetos (POO): Un Código más Inteligente

La POO es un paradigma para organizar el código. La idea es simple: en lugar de tener datos y funciones por separado, los agrupamos en "objetos" que tienen tanto los datos (atributos) como las funciones que operan sobre esos datos (métodos).

* **Clase**: Es el "plano" o la "receta". Define cómo serán los objetos.
* **Objeto (o instancia)**: Es la "casa" construida a partir del plano. Es una entidad real con la que interactuamos.
* **`__init__(self, ...)`**: El "constructor". Es un método especial que se ejecuta al crear un nuevo objeto para inicializar sus atributos.
* **`self`**: Se refiere al objeto mismo. Así es como el objeto accede a sus propios datos y métodos.

Vamos a crear una clase `ProcesadorTexto` que encapsule un texto y todas las operaciones que queremos hacer con él.

```python
# Definición ÚNICA de la clase. Esta celda es la referencia para todo el notebook.
class ProcesadorTexto:
    """Encapsula un texto y expone operaciones de análisis lingüístico básico."""

    def __init__(self, texto):
        # Nota sobre limpieza: strip() solo elimina caracteres en los extremos del string.
        # Para limpiar puntuación en todo el texto usamos replace(), igual que en limpiar_texto().
        print(f"Creando procesador para: '{texto[:40]}...'")
        self.texto_original = texto
        caracteres_a_eliminar = ['.', ',', '!', '?', ';', ':', '¿', '¡', '"', "'", '(', ')']
        texto_limpio = texto.lower()
        for char in caracteres_a_eliminar:
            texto_limpio = texto_limpio.replace(char, '')
        self.texto_limpio = texto_limpio
        self.palabras = self.texto_limpio.split()

    def contar_palabras(self):
        """Devuelve el total de palabras (con repeticiones)."""
        return len(self.palabras)

    def contar_palabras_unicas(self):
        """Devuelve la cantidad de palabras distintas. `set` elimina duplicados."""
        return len(set(self.palabras))

    def calcular_frecuencia(self):
        """Devuelve un diccionario con la frecuencia de cada palabra."""
        frecuencia = {}
        for palabra in self.palabras:
            frecuencia[palabra] = frecuencia.get(palabra, 0) + 1
        return frecuencia
```

```python
# --- Uso de la clase ---
# Creamos un objeto (una instancia) de nuestra clase
procesador = ProcesadorTexto("Python para NLP es genial porque Python es versátil.")

# Ahora podemos usar sus métodos
print(f"Total de palabras: {procesador.contar_palabras()}")
print(f"Palabras únicas: {procesador.contar_palabras_unicas()}")
print(f"Frecuencia de palabras: {procesador.calcular_frecuencia()}")
```

```python
#Mi versión de la función, con los pasos completos y la integración de la interfaz gráfica
import gradio as gr

# Copié la clase ProcesadorTexto acá y añadí el nuevo método calcular_longitud_media().
class ProcesadorTexto:
    """Encapsula un texto y expone operaciones de análisis lingüístico básico."""

    def __init__(self, texto):
        print(f"Creando procesador para: '{texto[:40]}...'")
        self.texto_original = texto
        caracteres_a_eliminar = ['.', ',', '!', '?', ';', ':', '¿', '¡', '"', "'", '(', ')']
        texto_limpio = texto.lower()
        for char in caracteres_a_eliminar:
            texto_limpio = texto_limpio.replace(char, '')
        self.texto_limpio = texto_limpio
        self.palabras = self.texto_limpio.split()

    def contar_palabras(self):
        return len(self.palabras)

    def contar_palabras_unicas(self):
        return len(set(self.palabras))

    def calcular_frecuencia(self):
        frecuencia = {}
        for palabra in self.palabras:
            frecuencia[palabra] = frecuencia.get(palabra, 0) + 1
        return frecuencia


def analizar_texto(texto):
    # Condición de seguridad
    if not texto.strip():
        return "Por favor, ingresá texto para analizar.", {}

    # Paso 1: Instancia del procesador
    procesador = ProcesadorTexto(texto)

    # Paso 2: Obtener datos usando los métodos de la clase
    total = procesador.contar_palabras()
    unicas = procesador.contar_palabras_unicas()
    frecuencia_diccionario = procesador.calcular_frecuencia()

    # Paso 3: Encontrar la palabra más frecuente
    #Ajusté esto porque me traía la primera de la lista de palabras como la más frecuente, si había igualdad de frecuencia.
    if frecuencia_diccionario:
        # 1. Encontramos cuál es el número máximo de repeticiones
        max_repeticiones = max(frecuencia_diccionario.values())
        
        # 2. Buscamos todas las palabras que tengan ese número de repeticiones
        palabras_top = [palabra for palabra, cuenta in frecuencia_diccionario.items() 
                        if cuenta == max_repeticiones]
        
        # 3. Decidimos qué mensaje mostrar
        if max_repeticiones == 1:
            resultado_frecuencia = "No hay palabras repetidas (todas aparecen una vez)."
        else:
            # Si hay varias, las unimos con comas
            formato_palabras = ", ".join([f"'{p}'" for p in palabras_top])
            resultado_frecuencia = f"{formato_palabras} ({max_repeticiones} veces)"
    else:
        resultado_frecuencia = "N/A"

    # Paso 4: Armar el string de resumen
    resumen = (
        f"--- Resumen del Análisis ---\n"
        f"Total de Palabras: {total}\n"
        f"Palabras Únicas: {unicas}\n"
        f"Palabra(s) Más Frecuente(s): {resultado_frecuencia}"
    )

   
#
    # Paso 5: Devolver el resumen y el diccionario
    return resumen, frecuencia_diccionario

    return "Función incompleta. Completá los pasos para ver el análisis.", {}

# --- PASO 2: Configurá la Interfaz de Gradio ---
demo = gr.Interface(
    fn=analizar_texto,

    inputs=gr.Textbox(
        lines=10,
        placeholder="Escribí o pegá un texto acá para analizarlo...",
        label="Texto a analizar"
    ),

    outputs=[
        gr.Textbox(label="Resumen del Análisis"),
        gr.JSON(label="Frecuencia de Palabras")
    ],

    title="Analizador de Texto",
    description="Introducí un texto y obtené un análisis básico de PLN.",
    flagging_mode="never"
)

# Descomentá la línea que corresponda a tu entorno:
demo.launch()                # En Google Colab (crea una URL pública temporal con share=True)
# demo.launch(inbrowser=True)  # En entorno local: abre el navegador automáticamente
# demo.launch(share=True)      # Crea un enlace público temporal (requiere internet)

# --- PASO 2: Interfaz de Gradio --- Esta es otra opción que me trajo Gemini

# app = gr.Interface(
  #  fn=analizar_texto,
   # inputs=gr.Textbox(lines=5, label="Ingresa tu texto aquí", placeholder="Había una vez..."),
    #outputs=[
     #   gr.Textbox(label="Resultado del Análisis"),
      #  gr.Label(label="Top de Frecuencias")
    #],
    #title="Analizador Lingüístico Pro",
    #description="Escribe un párrafo para analizar su estructura y frecuencia de palabras."
#)

# Lanzar la aplicación
#if __name__ == "__main__":
#    app.launch()
```

## Glosario Rápido — Clase 2

* **Método**: Una función que pertenece a un objeto (ej: `mi_texto.strip()`, `procesador.contar_palabras()`).
* **Clase**: El plano o plantilla para crear objetos. Define sus atributos y métodos.
* **Objeto / Instancia**: Una entidad concreta creada a partir de una clase.
* **Atributo**: Una variable que pertenece a un objeto (ej: `procesador.palabras`, `procesador.texto_original`).
* **`__init__(self, ...)`**: El constructor de la clase. Se ejecuta automáticamente al crear un objeto e inicializa sus atributos.
* **`self`**: Palabra clave que dentro de una clase se refiere a la instancia actual del objeto.
* **POO (Programación Orientada a Objetos)**: Paradigma que organiza el código agrupando datos y funciones en objetos.
* **Encapsulamiento**: La capacidad de ocultar detalles internos de implementación y exponer solo lo necesario a través de métodos.
* **Docstring**: Un string entre triple comillas (`"""..."""`), colocado al inicio de una función o clase, que describe su propósito.
* **Gradio**: Librería de Python que permite crear interfaces web interactivas para modelos y funciones de forma rápida.
